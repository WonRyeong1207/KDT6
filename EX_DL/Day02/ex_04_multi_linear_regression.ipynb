### [ 다중 선형 회귀 ]
---
- 행렬 곱셈 연산 또는 벡터 내적 연산(Dot Product)
    * 다중 x와 가중치 w 곱셈
        * [조건] 좌측에 있는 행렬의 열의 크기 == 우측에 있는 행렬의 행의 크기
        * [예시] (5 x 3) x (3 x 1)
        
            

            
    
    
    * 파이토치 Tensor에 메서드 제공 ==>  tensor.matmul()
    

```python
#모듈 로딩 ----------------------------------
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
# 랜덤 시드 설정 ------------------------------
torch.manual_seed(1)
```

```python
# 훈랸데이터 -----------------------------------
x_train  =  torch.FloatTensor([[73,  80,  75], 
                               [93,  88,  93], 
                               [89,  91,  80], 
                               [96,  98,  100],   
                               [73,  66,  70]])  
y_train  =  torch.FloatTensor([[152],  [185],  [180],  [196],  [142]])
```

```python
print(f'x_train :  {x_train.shape}, {x_train.ndim}차원')
print(f'y_train :  {y_train.shape}, {y_train.ndim}차원')
```

```python
# 모델 초기화 ---------------------------------------------
# 가중치와 편향 선언
W = torch.zeros((3, 1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)
```

```python
print(f'W ==> \n{W}\n\nb ==> {b}')
```

```python
# optimizer 설정
optimizer = optim.SGD([W, b], lr=1e-5)
```

```python
# 학습 진행 -------------------------------------------------
nb_epochs = 20
for epoch in range(nb_epochs + 1):

    # H(x) 계산
    # 편향 b는 브로드 캐스팅되어 각 샘플에 더해집니다.
    hypothesis = x_train.matmul(W) + b

    # cost 계산
    cost = torch.mean((hypothesis - y_train) ** 2)

    # cost로 H(x) 개선
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    print(f'Epoch {epoch:4d}/{nb_epochs} hypothesis: { hypothesis.squeeze().detach()} Cost: {cost.item():.6f}')
    
```

```python
# 예측 ------------------------------------------------------
# [[73,  80,  75]
pre_y=W[0]*73 + W[1]*80 + W[2]*75 + b

print(f'[73,  80,  75] pre_y => {pre_y}')
```
