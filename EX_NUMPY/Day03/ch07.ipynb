```python
fig = plt.figure(figsize=(8,6))
plt.title(' SIN Curve ')
plt.xlabel(" x axis ")
plt.ylabel(" x axis ")

plt.plot(x,y, color='0.75', linestyle='solid')
plt.plot(x+0.3, y,  color='#00ffff', linestyle='dashed')
plt.plot(x+0.6, y,  color=(0.35, 0.56, 0.65), linestyle='dashdot')
plt.plot(x+0.9, y,  color=(0.35, 0.56, 0.65, 0.50), linestyle='dashed')
plt.show()
```

- 2-2 함수를 위한 그래프 그리기

```python
# 함수 그래프 그리기
def spines() :
    ax = plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_position(('data',0))
    ax.spines['left'].set_position(('data',0))

    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
```

```python
x = np.arange(-10,13)
y = 3*x + 3
```

```python
plt.plot(x,y)
spines()
plt.show()
```

```python
y = -3*x + 3
```

```python
plt.plot(x,y)

spines()

plt.show()
```

```python
# 항등함수 알아보기
y = x
```

```python
plt. plot(x,y)
spines()
plt.show()
```

- 3-1 선형함수 알아보기

```python
# 항등함수 알아보기
def identity(x):
    return x
```

```python
x = np.arange(-10, 11)
y = identity(x)
x == y
```

```python
fig = plt.figure(figsize=(8,6))
plt.plot(x,y)
spines()
plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))
plt.plot(z, y)
spines()
plt.show()
```

- 3-2 비선형함수 알아보기

```python
# 계단함수 알아보기
def step(x):
    return np.where(x > 0, 1, 0)
```

```python
x = np.arange(-5, 5, 0.1)
y = step(x)
y
```

```python
plt.plot(x, y)
plt.grid()
spines()
plt.show()
```

```python
# 시그모이드 알아보기
def sigmoid(x) :
    return 1 /(1 + np.exp(-x))
```

```python
y = sigmoid(x)
x.shape, y.shape
```

```python
plt.plot(x, y)
plt.grid()
spines()
plt.show()
```

```python
# tanh 함수 알아보기
y = np.tanh(x)
```

```python
plt.plot(x, y)
plt.grid()
spines()
plt.show()
```

```python
# 렐루함수(ReLu function) 알아보기
def relu(x) :
    return np.maximum(0,x)
```

```python
y = relu(x)
```

```python
plt.plot(x, y)
plt.grid()
spines()
plt.show()
```

```python
# 소프트매스 함수 알아보기
def sotfmax(x):
    return np.exp(x) / np.sum(np.exp(x))
```

```python
x = np.arange(1, 5)
y = sotfmax(x)
y
```

```python
np.sum(y)
```

```python
ratio = y
labels = y
 
plt.pie(ratio, labels=labels, shadow=True, startangle=90)
plt.show()
```
