- 2-3 분산과 표준편차 이해하기

```python
# 분산과 표준편차 알아보기
a = np.array([[3,7,5],[8,4,3],[2,4,9]])
a
```

```python
np.var(a)
```

```python
m = np.mean(a)
m
```

```python
np.sum(np.power((a - m), 2)) / len(a.flatten())
```

```python
np.sqrt(np.var(a)), np.std(a)
```

```python
# 표본 분산 및 표준편차 알아보기
f = np.array([1,4,3,8,9,2,3])
np.var(f,ddof=1)
```

```python
m = np.mean(f)
m
```

```python
np.sum(np.power((f - m), 2))/ (len(f.flatten()) - 1)
```

```python
np.std(f, ddof=1), np.sqrt(np.var(f, ddof=1))
```

- 상관계수와 공분산 이해하기

```python
# 상관계수와 공분산 알아보기
a = np.array([5,6,7,8])
b = np.array([1,3,4, 5])
a, b
```

```python
np.corrcoef(a,b)    # 상관계수
```

```python
np.cov(a,b)     # 공분산
```

```python
ma = np.mean(a)
mb = np.mean(b)
np.sum((a -ma) * (b-mb)) / (len(a) - 1)
```

```python
np.std(a,ddof=1) * np.std(b, ddof=1)
```

```python
(np.sum((a -ma) * (b-mb)) / (len(a) - 1) ) / (np.std(a,ddof=1) * np.std(b, ddof=1))
```
