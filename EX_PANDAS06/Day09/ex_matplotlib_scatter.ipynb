### 산점도 그래프 - 버블
- 두개의 컬럼/변수의 관계를 시각화하여 표현
- 연속형/수치형 데이터에 사용함

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib import rc
```

```python
# 데이터 준비
file_path = '../data/auto_mpg.csv'
font_path = r'C:\Users\PC\Desktop\AI_빅데이터 전문가 양성과정 6기\KDT6\font\NanumMyeongjo.ttf'
```

```python
# 미리 지정하고 넘어가는 폰트

# 폰트 패밀리 이름 가져오기
font_name = fm.FontProperties(fname=font_path).get_name()

# 새로운 폰트 패밀리 이름 지정
rc('font',family=font_name)
```

```python
# 데이터 프레임
data_df = pd.read_csv(file_path)
data_df.info()
```

```python
# - 컬럼별 타입 확인
data_df.dtypes
```

```python
# 연비와 관련이 높은 컬럼 즉, 변수 확인
# - 상관계수 확인
corr_sr = data_df.corr(numeric_only=True)['mpg']
corr_sr
```

```python
# mpg와 weight의 상관관계 값
round(corr_sr['weight'], 2)
```

```python
# 산점도 그래프 그리기
plt.scatter(data_df['mpg'], data_df['weight'], label=f"corr: {round(corr_sr['weight'], 2)}")
plt.xlabel('mpg')
plt.ylabel('weight')
plt.legend()
plt.show()
```

```python
# mpg 컬럼과 다른 컬럼간의 관계 시각화
# - 그래프 담을 창 생성
fig = plt.figure(figsize=(14,7))

# - 그래프를 담을 공간 생성 : axes
axes = fig.subplots(2, 3, sharex=True)
for row in range(2):
    for col in range(3):
        col_name = corr_sr.index[row*3 + col + 1]
        axes[row, col].scatter(data_df['mpg'], data_df[col_name],
                               label=f"corr: {round(corr_sr[col_name], 2)}")

        axes[row, col].set_xlabel('mpg')
        axes[row, col].set_ylabel(corr_sr.index[row*3+col+1])
        axes[row, col].legend()
```

```python
# 위의 그래프를 스스로 그려보자.
# mpg와 다른 칼럼과의 상관관계
```

- [5] 연비와 무게, 실린더 수의 관계 시각화
    - 연비와 무게는 반비례 관계, 상관관계가 높음
    - 연비와 실린더 개수는 상관관계는 높은데 그래프 상으로는 알 수 없음
    - 그래서 연비, 무게, 실린더 3개로 그래프를 그리고 싶음

```python
# mpg와 weight 관계 산점도(scatter)

plt.scatter(data_df['mpg'], data_df['weight'], label=f'corr: {round(corr_sr["weight"], 4)}')

plt.title('연비와 무게')
plt.xlabel('연비')
plt.ylabel('무게')
plt.legend()

plt.show()
```

```python
# mpg와 weight 관계 산점도(scatter) 버블

plt.scatter(data_df['mpg'], data_df['weight'], label=f'corr: {round(corr_sr["weight"], 4)}',
            s=data_df['cylinders']*10, c=data_df['cylinders'])        # s : 크기, c : 색상

plt.title('연비와 무게')
plt.xlabel('연비')
plt.ylabel('무게')
plt.legend()

plt.show()
```

```python
# mpg와 weight 관계 산점도(scatter) 버블

plt.scatter(data_df['mpg'], data_df['weight'], label=f'corr: {round(corr_sr["weight"], 4)}',
            s=data_df['cylinders']*15, c=data_df['cylinders'],
            alpha=0.4, cmap='twilight')        # alpha : 투명도

plt.title('연비와 무게')
plt.xlabel('연비')
plt.ylabel('무게')
plt.legend()

plt.show()
```
