### 파이 그래프
- 컬럼/변수의 데이터의 분포 비율을 시각화

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib import rc
```

```python
# 데이터 준비
file_path = '../data/auto_mpg.csv'
font_path = r'C:\Users\PC\Desktop\AI_빅데이터 전문가 양성과정 6기\KDT6\font\NanumMyeongjo.ttf'
```

```python
# 미리 지정하고 넘어가는 폰트

# 폰트 패밀리 이름 가져오기
font_name = fm.FontProperties(fname=font_path).get_name()

# 새로운 폰트 패밀리 이름 지정
rc('font',family=font_name)
```

```python
# 데이터 프레임
data_df = pd.read_csv(file_path)
data_df.info()
```

```python
# - 컬럼별 타입 확인
data_df.dtypes
```

```python
# 특정 컬럼의 데이터 비율
# - 자동차 생산국가에 대한 비율
# 범주형으로 데이터 변화
data_df['origin'].astype('category')
```

```python
# - 자동차 생산국가에 대한 비율

# round() : 반올림해주는 내장함수
round(data_df['origin'].value_counts() / data_df['origin'].count() * 100, 2)
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - bar
origin_sr = data_df['origin'].unique()
x = data_df['origin'].unique()
h = round(data_df['origin'].value_counts() / data_df['origin'].count() * 100,2)


origin_bar = plt.bar(x, h, color = ['skyblue', 'orange', 'yellowgreen'], edgecolor=['cyan'])

plt.title('자동차 생산국가에 대한 비율')
plt.xlabel('생산국가')
plt.xticks(x, ['미국', '유럽', '일본'])     # 리스트로 범위를 지정해 줄 수 있다면 np.arange() 안써도 됨.
plt.ylabel('비율(%)')
plt.bar_label(origin_bar, h, label_type='center')     # bar 객체와 안에 들어갈 label 값, 라벨 위치

plt.show()
```

```python
# 자동차 생산국가에 대한 비율 시각화 - pie 그래프

origin_sr = data_df['origin'].unique()
x = data_df['origin'].unique()
h = round(data_df['origin'].value_counts() / data_df['origin'].count(),2)

plt.pie(h, labels=['미국', '유럽', '일본'], autopct='%.2f%%')   # autopct 스트링으로 줄때 '%~~'로 줘야하는 거였어.


plt.show()
```

```python
# 자동차 생산국가에 대한 비율 시각화 - pie 그래프

origin_sr = data_df['origin'].unique()
x = data_df['origin'].unique()
h = round(data_df['origin'].value_counts() / data_df['origin'].count(),2)

plt.pie(h, labels=['미국', '유럽', '일본'], autopct='%.2f%%',
        counterclock=False)   # True : 시계방향, False : 반시계방향


plt.show()
```

```python
# 자동차 생산국가에 대한 비율 시각화 - pie 그래프

origin_sr = data_df['origin'].unique()
x = data_df['origin'].unique()
h = round(data_df['origin'].value_counts() / data_df['origin'].count(),2)

plt.pie(h, labels=['미국', '유럽', '일본'], autopct='%.2f%%',
        startangle=-45)   # 기본은 None, 원하는 각도


plt.show()
```

```python
# 자동차 생산국가에 대한 비율 시각화 - pie 그래프

origin_sr = data_df['origin'].unique()
x = data_df['origin'].unique()
h = round(data_df['origin'].value_counts() / data_df['origin'].count(),2)

plt.pie(h, labels=['미국', '유럽', '일본'], autopct='%.2f%%',
        explode=[0, 0, 0.1])   # 그래프 찢기


plt.show()
```

```python
# 자동차 생산국가에 대한 비율 시각화 - pie 그래프

origin_sr = data_df['origin'].unique()
x = data_df['origin'].unique()
h = round(data_df['origin'].value_counts() / data_df['origin'].count(),2)

plt.pie(h, labels=['미국', '유럽', '일본'], autopct='%.2f%%',
        wedgeprops={'linewidth':2, 'width':0.7, 'edgecolor':'cyan'})   # 도넛 만들기, 색을 흰색으로 주면 간격을 떨어뜨린 것 처럼 보임.
                                                                       # width는 그래프의 넓이이기때문에 1이하의 소수점으로 값을 줘야 떨어짐


plt.show()
```
