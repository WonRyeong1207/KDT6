### 멀티인덱스

```python
import pandas as pd
```

```python
file_path = '../data/test_data_3.csv'
```

```python
df1 = pd.read_csv(file_path)
df1
```

```python
# 대구 관련 데이터 추출
df1[(df1['col1'] == '대구')]
```

```python
# index와 col1 컬럼을 핸 인덱스 설정
df2 = df1.set_index(['index', 'col1'])
df2
```

```python
# 인덱스 속성
df2.index, df2.values
```

```python
# 2020년도 대구 데이터 추출
df2.loc[('2020', '대구')]
```

```python
# 2021년도 대구
df2.loc[('2021', '대구')]
```

```python
df2.loc['2020' ,'대구'] = [9, 9]
df2
```

```python
df2.loc[('2021', '대구')] = [3, 3]
df2
```

```python
# 행 추가 => (년도, 지역명)
# df2.loc[('2022', '부산')] = 0       # 행, 열 로 인식해서 열추가로 가버림
df2.loc[(2020, '부산'), :] = 0      # 핼과 열을 확실하게 구분해야함.
df2
```

```python
df2.loc[('2022', '마산'), :] = [3, 4]
```

```python
df2
```

```python
df2.loc[('2022', '울산'), :] = [pd.Series({'col2':8, 'col3':8})]
```

```python
df2
```

```python
df2.loc[('2022', '경산'), :] = [pd.Series({'col2':3, 'col3':2})]
df2
```
