### CSV 데이터 파일 다루기
- auto_mpg 데이터 파일 다루기

```python
import pandas as pd
```

```python
file_path = '../data/auto_mpg.csv'
```

```python
mpg_df = pd.read_csv(file_path) # nrow, usecols

mpg_df
```

```python
print(f"데이터 행수 : {mpg_df.shape[0]}, 열 개수 : {mpg_df.shape[1]}")
print(f"데이터의 칼럼 : {mpg_df.columns}")
print(f"데이터의 인덱스 : {mpg_df.index}")
print(f"데이터의 타입 : \n{mpg_df.dtypes}")
```

```python
# 70년대 자동차 데이터 추출
bool_idx = (70 <= mpg_df['model year']) & (mpg_df['model year'] < 80)
mpg_df[bool_idx]
```

```python
# 실린더 개수가 4개이하인 자동차 데이터 추출
bool_idx = mpg_df['cylinders'] <= 4
mpg_df[bool_idx]
```

```python
# 일본차가 아닌 다머지 자동차 데이터 추출
bool_idx = mpg_df['origin'] != 3
mpg_df[bool_idx]
```

```python
# 80년대 유럽 자동차 중에서 실린더 개수가 6개 이상인 자동차 데이터만 추출
# mpg_df['model year'] & 80 < 10
bool_idx = (80 <= mpg_df['model year']) &(mpg_df['model year'] < 90) & (mpg_df['origin'] == 2) & (mpg_df['cylinders'] >= 6)
mpg_df[bool_idx]
```
