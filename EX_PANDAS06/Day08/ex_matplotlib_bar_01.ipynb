- 기본확인에서 파악된 내용들 처리

```python
# - 불필요 데이터 처리
data_df_1 = data_df.drop(index=0)
data_df_1.head()
```

```python
# - 결측치 처리
data_df_1.isna().sum()
```

```python
# -1984, 1985, 1986 칼럼만 추출
# data_df_1.columns
a = data_df_1[['1984', '1985', '1986']].isna()
a[a.sum(axis=1) > 0]
```

```python
# 전출지 결측지 채우기
data_df_1['전출지별'] = data_df_1['전출지별'].ffill()
data_df_1['전출지별']
```

```python
# 값이 없는 결측치 채우기
data_df_1[['1984', '1985', '1986']] = data_df_1[['1984', '1985', '1986']].fillna(0) # 컬럼을 들고 오니까 안뜬다.
```

- 결측치부분 추가 설명

```python
df2 = pd.DataFrame([[1, 2, pd.NA], [pd.NA, pd.NA, 2], [pd.NA, 3, 3]])
df2
```

```python
# 0번 컬럼을 이전 값으로 채우기
df2[0] = df2[0].ffill()
df2
```

```python
# 이상치 처리
data_df_2 = data_df_1.replace('-', 0)
data_df_2.head()
```

```python
# 자료형 변환
# 컬럼별로 데이터형을 다르게 해야하는 경우 이런 방법으로 처리 해줄 수 있음
d_ = dict(zip(data_df_2.columns[2:].to_list(), ['int']*(data_df_2.shape[-1] -2)))   # 저체 컬럼 수에서 2개 빼고 int
data_df_2 = data_df_2.astype(d_)
```

[5] 분석 : 대구에서 전지역 전입 인구수 분석

```python
# 그래프 그리기 전에 한글폰트 설정

# 사용할 폰트 파일
font_path = r'C:\Users\PC\Desktop\AI_빅데이터 전문가 양성과정 6기\KDT6\font\NanumMyeongjo.ttf'

# 폰트 패밀리 이름 가져오기
font_name = fm.FontProperties(fname=font_path).get_name()

# 새로운 폰트 패밀리 이름 지정
rc('font',family=font_name)
```

```python
deagu_df_1
```

```python
# 그래프 제목 : 대구 타지역 이동 인구
# X축 : deagu_df_1.index
# Y축 : 인구수 합계
# 그래프 종류 : 막대그래프

# 데이터 정리
x_data = deagu_df_1.index.to_list()

y_data = []
for idx in range(len(deagu_df_1.index)):
    y_data.append(deagu_df_1.iloc[idx, :].sum())

# 그래프 그리기
plt.bar(x_data, y_data, width=0.3)
plt.title('대구 -> 타지역 이동 인구수')
plt.xlabel('행정구역')
plt.xticks(np.arange(0,16,1), rotation=45)
plt.ylabel('인구수(명)')
plt.tight_layout()
plt.show()
```

```python
# 그래프 제목 : 대구 타지역 이동 인구 (경북제외)
# X축 : deagu_df_1.index
# Y축 : 인구수 합계
# 그래프 종류 : 막대그래프

# 데이터 정리
deagu_df_1_2 = deagu_df_1.drop(index='경상북도')

x_data = deagu_df_1_2.index.to_list()

y_data = []
for idx in range(len(deagu_df_1_2.index)):
    y_data.append(deagu_df_1_2.iloc[idx, :].sum())

# 그래프 그리기
plt.bar(x_data, y_data, width=0.3)
plt.title('대구 -> 타지역 이동 인구수 (경상북도 제외)')
plt.xlabel('행정구역')
plt.xticks(np.arange(0,15,1), rotation=45)
plt.ylabel('인구수(명)')
plt.tight_layout()
plt.show()
```

```python
# 그래프 제목 : 대구 타지역 이동 인구 (경북, 서울, 경기도, 경남 제외)
# X축 : deagu_df_1.index
# Y축 : 인구수 합계
# 그래프 종류 : 막대그래프

# 데이터 정리
deagu_df_1_3 = deagu_df_1_2.drop(index=['경상남도', '서울특별시', '경기도', '경상남도'])

x_data = deagu_df_1_3.index.to_list()

y_data = []
for idx in range(len(deagu_df_1_3.index)):
    y_data.append(deagu_df_1_3.iloc[idx, :].sum())

# 그래프 그리기
plt.bar(x_data, y_data, width=0.3)
plt.title('대구 -> 타지역 이동 인구수 (경상북도, 서울특별시, 경기도, 경상남도 제외)')
plt.xlabel('행정구역')
plt.xticks(np.arange(0,12,1), rotation=45)
plt.ylabel('인구수(명)')
plt.tight_layout()
plt.show()
```

- 1970 ~ 2017년 사이 대구에서 타지역 이동 인구수는 경상북도, 서울특별시, 경기도, 경상남도, 부산광역시 순으로 높은 걸로 나타났다.
- 세종특별자치시, 광주광역시, 전라남도, 전라북도 순으로 낮은 것을 볼 수 있다.

```python
# 그래프 제목 : 대구 타지역 이동 인구(년간)
# X축 : deagu_df_1.index
# Y축 : 인구수 합계 연간 합계
# 그래프 종류 : 막대그래프
# 범례 : deagu_df_1.index

# 데이터 정리
x_data = deagu_df_1.columns.to_list()
y_data = deagu_df_1.sum().to_list()
label = deagu_df_1.index.to_list()


# 그래프 그리기
plt.bar(x_data, y_data, width=0.3)
plt.title('대구 -> 타지역 이동 인구수 (1970 ~ 2017)')
plt.xlabel('년도')
plt.xticks(np.arange(0,48,5), ['1970', '1975', '1980', '1985', '1990', '1995', '2000', '2005', '2010', '2015'], rotation=45)
plt.ylabel('인구수(명)')
plt.tight_layout()
plt.show()
```
