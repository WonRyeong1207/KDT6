### DF 연결/통합
- 통합
    - DF에서 특정칼럼을 기준으로 통합해주는 기능
    - pandas.merge()
    - DataFrame.merge()
    - DataFrame.join()

```python
import pandas as pd
```

```python
file_path_1 = r'C:\Users\PC\Desktop\AI_빅데이터 전문가 양성과정 6기\KDT6\EX_PANDAS06\data\stock_price.xlsx'
file_path_2 = r'C:\Users\PC\Desktop\AI_빅데이터 전문가 양성과정 6기\KDT6\EX_PANDAS06\data\stock_valuation.xlsx'
```

```python
price_df = pd.read_excel(file_path_1)
val_df = pd.read_excel(file_path_2)
```

```python
print(price_df.head(2), price_df.tail(2), sep='\n\n')
```

```python
# DF 합치기 => pandas.merge() 함수
stock_df = pd.merge(price_df, val_df)   # 기본은 동일한 데이터 기준으로 합쳐짐
stock_df.info()
```

```python
stock_df
```

```python
# - DF 별 특정 컬럼 지정
stock_df = pd.merge(price_df, val_df, left_on='stock_name', right_on='name')
```

```python
stock_df
```

```python
stock_df = pd.merge(price_df, val_df,how='outer')
```

```python
stock_df
```

```python
stock_df = pd.merge(price_df, val_df, how='right')
stock_df
```

```python
stock_df = pd.merge(price_df, val_df, how='left')
stock_df
```

```python
stock_df = pd.merge(price_df, val_df, how='cross')  # 강제로 데이터를 늘일때 사용한다고 함
stock_df
```

```python
# DF 합치기 => DataFrame.join() 메서드
price_df.join(val_df, lsuffix='_price', rsuffix='_val')
```
