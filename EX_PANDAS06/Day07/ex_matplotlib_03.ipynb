### MatPlotLib 활용 데이터 시각화
-  X축과 Y축 다양한 처리

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 그래프 그리기 - normal
y_data = [x for x in range(1, 10, 2)]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']

plt.plot(x_data, y_data, marker='o')
plt.show()
```

```python
# 그래프 그리기 - x축과 y축 범위 설정
y_data = [x for x in range(1, 10, 2)]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']

plt.plot(x_data, y_data, marker='o')
plt.ylim(bottom=0, top=15)
plt.xlim(left='2023-12-01', right='2024-06-01') # 무조건 숫자로 줘야함. 숫자로 주고 나중에 이름을 줘야함.
plt.show()
```

```python
# 그래프 그리기 - x축과 y축의 눈금(tick) 설정
y_data = [x for x in range(1, 10, 2)]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']

plt.plot(x_data, y_data, marker='o')
plt.xticks(rotation=45)
plt.show()
```

```python
# 그래프 그리기 - 그래프 출력창, figure 크기 설정
y_data = [x for x in range(1, 10, 2)]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']


plt.figure(figsize=[5, 5])
plt.plot(x_data, y_data, marker='o')
plt.show()
```

```python
# 그래프 그리기 - 그래프 출력 창 figure 자동맞춤
y_data = [x for x in range(1, 10, 2)]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']


# plt.figure(figsize=[5, 5])
plt.plot(x_data, y_data, marker='o')
plt.ylim(bottom=0, top=15)
plt.tight_layout()      # 모든 내용이 출력될 수 있도록 자동 맞춤 설정
plt.show()
```

```python
# 그래프 그리기 - 데이터다 중심.
y_data = [x for x in range(1, 10, 2)]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']


plt.plot(x_data, marker='o')
plt.ylim(bottom=0, top=15)
plt.xlim(left=0, right=7)
plt.xticks([1,2,3,4,5])
plt.show()
```

```python
# 그래프 그리기 - 데이터다 중심.
y_data = [x for x in range(1, 10, 2)]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']


plt.plot(y_data, marker='o')
plt.ylim(bottom=1, top=15)
plt.xlim(left=1, right=7)
plt.xticks([0,1,2,3,4,5,6,7])   # 리미트와 눈급을 같이 줄 수 있음.
plt.show()
```

```python
# 그래프 그리기 - 데이터다 중심.
y_data = [x for x in range(1, 10, 2)]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']


plt.plot(y_data, marker='o')
plt.ylim(bottom=-1, top=10)
plt.xlim(left=-1, right=6)
plt.xticks([0, 2, 4, 6], ['A', 'B', 'C', 'D'], fontsize = 12, fontweight='bold')   # 리미트와 눈급을 같이 줄 수 있음. 두번째는 라벨값.
plt.show()
```

```python
# 그래프 그리기 - 데이터다 중심.
y_data = [x for x in range(1, 10, 2)]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']


plt.plot(y_data, marker='o')
plt.ylim(bottom=-1, top=10)
plt.xlim(left=-1, right=6)
font_config = {'fontsize':22, 'fontweight':'bold'}
plt.xticks([0, 2, 4, 6], ['A', 'B', 'C', 'D'], **font_config)   # **dict : 키워드 파라미터로 사용하겠다
plt.show()
```
