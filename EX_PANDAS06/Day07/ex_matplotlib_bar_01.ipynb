### Matplotlib 활용 시각화 - 막대 그래프
- 가장 많이 사용되는 그래프
- 데이터를 비교할때 사용

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 데이터 준비
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']
```

```python
# 데이터 시각화
plt.bar(fruits, counts, width=0.3)
plt.show()
```

```python
# 데이터 시각화
plt.bar(fruits, counts, width=0.3, color=bar_colors)
plt.show()
```

```python
# 데이터 시각화
plt.bar(fruits, counts, width=0.7, color=bar_colors, label=bar_labels)
plt.legend()    # 라벨은 범례와 관계 있음
plt.show()
```

```python
# 수평맏개 그래프
plt.barh(fruits, counts, height=0.7, color=bar_colors, label=bar_labels)
plt.legend()    # 라벨은 범례와 관계 있음
plt.show()
```

```python
# 데이터 시각화 - 막대 + 선 그래프
plt.bar(fruits, counts, width=0.3, color=bar_colors, label=bar_labels)
plt.plot(counts, 'bo--')
plt.legend()    # 라벨은 범례와 관계 있음
plt.show()
```
