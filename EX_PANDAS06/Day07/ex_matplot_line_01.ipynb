### Matplotlib 활용한 시각화
- matplotlib.pyplot 모듈을 사용
- 다양한 그래프 관련 함수들 & 클래스들 존재

```python
import pandas as pd
import matplotlib.pyplot as plt     # 패키지명.모듈명
```

```python
plt.plot([1, 2, 3, 4])  # 기본이 y축 데이터, 선이 기본
plt.xlabel('data')
plt.ylabel('result')
plt.title('y = f(x)')
plt.show()      # 그치 이거 없으면 plt가 아니지. 맨날 이거로 보이는 거라고 배웠는데
```

```python
# x축 데이터를 줄거임
plt.plot([10, 20, 30], [5, 10, 15])
plt.show()
```

```python
# fomat string => fmf
# fmf 포맷스트링 : 그래프의 색상, 두계, 스타일 등등 지정
plt.plot([10, 20, 30], [5, 10, 15], 'r--')
plt.show()
```

```python
# 만나는 지점에 대해서 뭔가를 줄 수 있음.
plt.plot([10, 20, 30], [5, 10, 15], 'ro--')
plt.show()
```

```python
# o, ^, <, > 등등 색상은 컬러 이름으로 검정은 k
plt.plot([10, 20, 30], [5, 10, 15], 'y^--')
plt.show()
```

```python
# 속성으로 설정, 내가 배웠던 설정하는 방법.
plt.plot([10, 20, 30], [5, 10, 15], color='red', marker='o', linestyle='dashed')
plt.show()
```

```python
# 속성으로 설정, 내가 배웠던 설정하는 방법. 속성의 종류는 다앙함
plt.plot([10, 20, 30], [5, 10, 15], color='red', marker='o', linestyle='dashed',
         markersize=10, linewidth=4)
plt.show()
```

```python
# 속성이 너무 기니까 줄여서 씀
# mec='blue', mew=5 : marker의 설정. 전부 외울수는 없으니까 홈페이지에서
# 원하는 속성을 보고 찾아서 써야함.
plt.plot([10, 20, 30], [5, 10, 15], color='red', marker='o', linestyle='dashed',
         markersize=10, linewidth=4, mec='blue', mew=5)
```

```python
# 그래프 여러개 그리기
plt.plot([10, 20, 30], [15, 25, 35], 'bo-')
plt.plot([10, 20, 30], [5, 15, 20], 'ro--')
plt.plot([10, 20, 30], [12, 24, 36] ,'y^--')
plt.show()
```

```python
plt.plot([10, 20, 30], [15, 25, 35], 'bo-',
         [10, 20, 30], [5, 15, 20], 'ro--',
         [10, 20, 30], [12, 24, 36] ,'y^--')
plt.show()      # 한번에 그릴수도 있음.
```

```python
# 그래프의 부가적인 기능 추가
# - 그래프 제목, xy축 라벵, 범례

plt.plot([10, 20, 30], [15, 25, 35], 'bo-')
plt.plot([10, 20, 30], [5, 15, 20], 'ro--')
plt.plot([10, 20, 30], [12, 24, 10] ,'y^--')
plt.title('[Mplib Line 2D]')
plt.xlabel('data')
plt.ylabel('result')
plt.legend('value')     # 이렇게 하는것보다 값을 지정해주는 방법이 있었을 텐데..
plt.show()
```

```python
# 그래프의 부가적인 기능 추가
# - 그래프 제목, xy축 라벵, 범례

plt.plot([10, 20, 30], [15, 25, 35], 'bo-', label='data1')
plt.plot([10, 20, 30], [5, 15, 20], 'ro--', label='data2')
plt.plot([10, 20, 30], [12, 24, 10] ,'y^--', label='data3')
plt.title('[Mplib Line 2D]')
plt.xlabel('data')
plt.ylabel('result')
plt.legend()
plt.show()
```

```python
# 그래프의 부가적인 기능 추가
# - 그래프 제목, xy축 라벵, 범례
# plt.plot()에서 label 로 줄수 있고, plt.legend에서 loc 속성 o로 위치 지정 가능

plt.plot([10, 20, 30], [15, 25, 35], 'bo-', label='data1')
plt.plot([10, 20, 30], [5, 15, 20], 'ro--', label='data2')
plt.plot([10, 20, 30], [12, 24, 10] ,'y^--', label='data3')
plt.title('[Mplib Line 2D]')
plt.xlabel('data')
plt.ylabel('result')
plt.legend(loc='upper center')
plt.show()
```
