### 남북한 전력파일 실습
- 신재생 에너지 항목 북한 추가
- 없는 값은 '-' ==> 0으로 변환
- 2016년 열 옆에 항복별 합계 칼럼 추가, 추가 값은 0으로 추가
- 값은 for로 행별 합계, 함수 만들어서 사용

```python
import pandas as pd
```

```python
data = pd.read_excel('남북한발전전력량.xlsx', )
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.loc[9] = ['NaN', '신재생', '-','-','-','-','-', '-','-','-','-','-', '-','-','-','-','-', '-','-','-','-','-', '-','-','-','-','-', '-','-']
```

```python
df
```

```python
df.replace({'-':0}, inplace=True) # 이거 없어지는 메서드래요...
# import 하고 pd.set_option('future.no_silent_downcasting', 'True') 선언을 하라는 건가..
# 없어지만 반목문 돌려야 하나? 뭐지?
```

```python
df['항목별 합계'] = 0
```

```python
df
```

```python
# 항목별 합계를 구하는 함수
def col_sum(range_data):
    total = 0
    for i in range_data:
        total = total + i
    return total
```

```python
total = col_sum(df.iloc[0, 2:29])
```

```python
print(total)
```

```python
df.iloc[0, 29] = total
```

```python
df
```

```python
total = []
for r in range(len(df.index)):
    total.append(col_sum(df.iloc[r, 2:29]))
print(total)
```

```python
df['항목별 합계'] = total
```

```python
df
```
