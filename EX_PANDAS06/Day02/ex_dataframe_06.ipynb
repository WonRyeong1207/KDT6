### DataFrame 행/열 추가
- 행 방향 추가
- 열 방향 추가

```python
import pandas as pd
```

```python
data = {'name':['홍길동', '마징가', '원더우먼'],
        'job':['의적', '로봇', '영웅'],
        'gender':['남', '남', '여'],
        'age':[100, 200, 300],
        'height':[180, 300, 200]}
```

```python
dataDF = pd.DataFrame(data)
```

```python
dataDF
```

- 실습 :  1번 행을 추가 : 변수명.loc['추가할 행이름] = 새로운 데이터
    * 변수명.iloc[] = ['']는 에러

```python
# 1개 행 추가  => 데이터 수가 열 갯수와 동일한 경우
dataDF.loc[3] = ['베트맨', '박쥐', '남', 40, 190]
```

```python
dataDF
```

```python
# 1개 행 추가 =>  데이터 1개  => 모든 영역에 동일한 데이터 적용
dataDF.loc[5] = 100
```

```python
dataDF.loc[4] = ['스파이더맨', '거미', '남', 18, 185]
```

```python
dataDF
```

- 실습 행 추가 : 변수명.loc['추가할 행 이름] = 데이터
    * 변수명.iloc[] XX
    * 데이터수 == 행 수 경우, 열 추가 가능
    * 데이터 수 == 1개 경우, 열 추가 가능
