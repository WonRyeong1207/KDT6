```python
with open(DATA_PATH+file_list[0], mode='rb') as f:
    # 파일의 헤더정보 읽기
    _, cnt = struct.unpack('>II', f.read(8))
    rows, cols = struct.unpack('>II', f.read(8))
    
    # 이미지 raw 데이터 읽기: 1개
    pixels = f.read(rows*cols)
    print(cnt)
    print(rows*cols)
    print(pixels)   # type byte
```

```python
with open(DATA_PATH+file_list[1], mode='rb') as f:
    # 라벨
    _, cnt = struct.unpack('>II', f.read(8))
    
    label = f.read(cnt)
    print(cnt)
    print(label)
```

```python
def make_data_file(img_file, label_file, datasets_file):
    
    with open(label_file, mode='rb') as l_f:
        _, cnt = struct.unpack('>II', l_f.read(8))
        with open(img_file, mode='rb') as f_f:
            _, _ = struct.unpack('>II', f_f.read(8))
            rows, cols = struct.unpack('>II', f_f.read(8))
            colnames = [f'pixel{x}' for x in range(1, rows*cols+1)]
            colnames.insert(0, 'label')
            with open(datasets_file, mode='w') as d_f:
                d_f.write(','.join(colnames)+'\n')
                for i in range(cnt):
                    pixels = f_f.read(rows*cols)
                    # sdata = list(map(lambda n:int(n), pixels))
                    sdata = list(map(lambda n:str(n), pixels))
                    # print(sdata)
                    
                    label = struct.unpack('B', l_f.read(1))[0]
                    # print(label)
                    
                    # sdata.append(str(label))
                    sdata.insert(0, str(label))
                    # print(','.join(sdata))
                    # print(sdata)
                    img_label = ','.join(sdata)
                    d_f.write(img_label+'\n')
```

```python
make_data_file(DATA_PATH+file_list[0], DATA_PATH+file_list[1], '../data/MNIST/mnist_test.csv')
make_data_file(DATA_PATH+file_list[2], DATA_PATH+file_list[3], '../data/MNIST/mnist_train.csv')
```
