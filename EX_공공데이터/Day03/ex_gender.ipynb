```python
import matplotlib.pyplot as plt
import koreanize_matplotlib
import numpy as np
import csv
```

```python
def print_population(population):
    '''
    특정 지역의 인구 현황을 화면에 출력함
    '''
    
    for i in range(len(population)):
        print(f"{i:3d}세: {population[i]:4d}명", end='  ')
        if (i+1)%10 == 0:
            print()
    print()
```

```python
def draw_gender_population(title, male_num_list, female_num_list):
    plt.bar(range(len(male_num_list)), male_num_list, label='남성')
    plt.bar(range(len(female_num_list)), female_num_list, label='여성')
    plt.title(title + ' 성별 인구 비율')
    plt.legend()
    plt.show()
```

```python
def calculate_population():
    f = open('../data/gender.csv', encoding='euc_kr')
    data = csv.reader(f)
    male_num_list = []
    female_num_list = []
    
    distict = input('시군구 이름을 입력하세요: ')
    for row in data:
        if distict in row[0]:
            for male in row[106:207]:
                male = male.replace(',', '')
                male_num_list.append(-int(male))
            for female in row[209:310]:
                female = female.replace(',', '')
                female_num_list.append(int(female))
    f.close()
    
    print(f"남성 총 인구수: {sum(male_num_list):,}")
    print_population(male_num_list)
    print('--------------------------------------------')
    print(f"여성 총 인구수: {sum(female_num_list):,}")
    print_population(female_num_list)
    
    draw_gender_population(distict, male_num_list, female_num_list)
```

```python
calculate_population()
```
