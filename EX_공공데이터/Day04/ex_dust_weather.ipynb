미세먼지 + 기온데이터

```python
# 필요한 라이브러리
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt
import koreanize_matplotlib
import numpy as np
```

```python
dust_df = pd.read_excel('../data/dust.xlsx')

dust_df.head()
```

```python
print(tabulate(dust_df.head(), headers='keys', tablefmt='pretty'))
```

```python
dust_df.info()
```

```python
dust_df.rename(columns={'날짜':'date', '아황산가스':'SO2',
                        '일산화탄소':'CO', '오존':'O3',
                        '이산화질소':'NO2'}, inplace=True)

print(tabulate(dust_df.head(), headers='keys', tablefmt='psql'))
```

```python
dust_df['date'] = dust_df['date'].str[:10]
print(tabulate(dust_df.head(), headers='keys', tablefmt='psql'))
```

```python
dust_df['date'] = pd.to_datetime(dust_df['date'])
print(dust_df.dtypes)
```

```python
dust_df['year'] = dust_df['date'].dt.year
dust_df['month'] = dust_df['date'].dt.month
dust_df['day'] = dust_df['date'].dt.day
print(dust_df.columns)
```

```python
# 컬럼 정렬
dust = dust_df[['date', 'year', 'month', 'day', 'SO2', 'CO', 'O3', 'NO2', 'PM10', 'PM2.5']]
print(dust.columns)
```

```python
# 결측치 확인
print("결측치 확인하기")
dust.isna().sum()
```

```python
print("결측치를 포함한 데이터 출력")
dust[dust.isna().any(axis=1)]   # any or같은 느낌
```

```python
# 결측치 채우기
dust = dust.ffill()
dust.isna().sum()
```

```python
dust.iloc[132:134]  # 값이 채워진 것을 볼 수 있음.
```

```python
# 미세먼지처럼 날씨정보도 들고오기
weather_df = pd.read_excel('../data/weather.xlsx')
print(tabulate(weather_df.head(), headers='keys', tablefmt='psql'))
```

```python
weather_df.info()
```

```python
weather = weather_df.drop(['지점', '지점명'], axis=1)
weather.columns = ['date', 'temp', 'wind', 'rain', 'humidity']

print(tabulate(weather.head(), headers='keys', tablefmt='pretty'))
```

```python
weather['date'] = pd.to_datetime(weather['date'].dt.date)
# 뒤에 나오는 시각 정보를 삭제
```

```python
weather.info()
```

```python
weather.head()
```

```python
# 결측치 확인
print("날씨 데이터 결측치 개수 확인하기")
weather.isna().sum()
```

```python
print('날씨 데이터에서 결측치를 포함하는 행 출력')
weather[weather.isna().any(axis=1)]
```

```python
# 결측치 채우기
weather = weather.ffill()
weather.isna().sum()
```

```python
weather.iloc[[369, 565, 742]]   # replace 이후라서 0.01로 표시됨.
```

```python
# 강수량 데이터 변경하기
# - 기상청에서는 0.1 단위로 강수량을 측정
# - 강수량이 0.1 이하면 0으로 표시
# - 0을 0.0로 변경

print("강수량이 0인 항목을 0.01로 변경")
weather['rain'] = weather['rain'].replace(0, 0.01)
weather['rain'].value_counts()
```

```python
# 미세먼지와 날씨 데이터를 합치기 위한 작업
print('dust, weather의 크기 확인')
print(f"dust shape: {dust.shape}")
print(f"weather shape: {weather.shape}")
```

```python
# 미세먼지 데이터에서 공통 내용이 아닌 행 삭제
# - 행의 수가 다르기 때문에 맞춰주기 위함
dust.iloc[740:]
```

```python
weather.iloc[740:]
```

```python
# dust의 마지막은 23-12-31 00시인데 weather에서는 24-01-01 00로 표기함
dust = dust.drop(index=743)
print(dust.shape)
```

```python
# 데이터 프레임 병합하기
# merge: 왼쪽 데이터프레임을 기준으로 병합

print('dust, weather 데이터 프레임 merge')
merged_df = pd.merge(dust, weather, on='date')
merged_df.head()
```

```python
print(tabulate(merged_df.head(), headers='keys', tablefmt='psql'))
```

- 데이터 전처리 끝, 데이터분석 시작

```python
# 상관관계 분석
# - 모든 요소별 상관관계 확인

# 전체 행과 컬럼 다 보이게 하기
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# 상관관계 출력
print(merged_df.corr())
```

```python
# 미세먼지(PM10)과 상관관계
print('미세먼지(PM10)과 상관관계 분석')
corr = merged_df.corr()

corr['PM10'].sort_values(ascending=False)   # 내림차순 정렬
```

```python
# 각 요소별 히스토그램_판다스에서 지원하는 기능
merged_df.hist(column=['SO2', 'CO', 'O3', 'NO2', 'PM10', 'PM2.5', 'temp', 'wind',
                       'rain', 'humidity'], bins=50, figsize=(20,15))
plt.show()
```

```python
# 그래프를 이쁘게 그리게 위해 seaborn
import seaborn as sns
```

```python
# 날짜별 PM10 농도 막대 그래프

plt.figure(figsize=(15,10))
daygraph = sns.barplot(x='day', y='PM10', data=merged_df)
plt.title('날짜별 PM10 농도')
plt.show()
```

```python
# 상과관계 히트맵

plt.figure(figsize=(15, 10))
sns.heatmap(data=corr, annot=True, fmt='.2f', cmap='hot')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(data=corr, annot=True, fmt='.2f')
plt.show()
```

```python
# PM10과 PM2.5의 산점도
plt.figure(figsize=(15,10))
x = merged_df['PM10']
y = merged_df['PM2.5']

plt.plot(x, y, marker='o', linestyle='none', color='red', alpha=0.5)
plt.title('PM10 vs PM2.5')
plt.xlabel('PM10')
plt.ylabel('PM2.5')
plt.show()
```
