### Nature Language Processing Preprocessing -> Tokenization
- Tokenization: 의미를 가진 최소 단위(형태소)로 나누는 작업
    - word tokenize
    - sent torkenize
    - byte torkenize
    - 종류는 다양함 'https://www.nltk.org/api/nltk.html' 여기서 필요한거 찾아서 보셈. 사이트 좀 별로임. 보기가 ㅈfkf이네

```python
from nltk.tokenize import word_tokenize, sent_tokenize, wordpunct_tokenize
```

```python
text = 'He is gone.'

print(f"word tokenize: {word_tokenize(text)}")
print(f"sent tokenize: {sent_tokenize(text)}")
print(f"wordpunct tokenize: {wordpunct_tokenize(text)}") # 단어를 구두문자 기준으로 분리
```

- Pos-Tagging: Part Of Speech Tagging

```python
from nltk.tag import pos_tag
```

```python
one_text = "I am actively looking for Ph.D. students. and you are a Ph.D student."

print(f"word tokenization: {word_tokenize(one_text)}")
print(f"wordpunct tokenizaton: {wordpunct_tokenize(one_text)}")
print(f"pos tagging: {pos_tag(word_tokenize(one_text))}")   # 해주는 것 마다 다름, 문법적으로 표현 하는 것들이
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
# lemming [default: n]
words = ['policy', 'doing', 'organization', 'have', 'going', 'love', 'lives', 'fly', 'dies', 'watched', 'has', 'starting']
lemming = [WordNetLemmatizer().lemmatize(word) for word in words]

print(f"words: {words}")
print(f"setmming: {lemming}")
```

```python
from nltk.stem import PorterStemmer, LancasterStemmer
```

```python
# stemming
porter_stemming = [PorterStemmer().stem(word) for word in words]
lancaster_stemming = [LancasterStemmer().stem(word) for word in words]
# 같은 stem 추출이지만 algorithm이 다름. 궁금하면 document 봐야하고 봐도 모르겠으면 source code 뜯어봐야함

print(f"words: {words}")
print(f"porter stemming: {porter_stemming}")    
print(f"lancaster stemming: {lancaster_stemming}")
# 만약 한글을 한다면 종유도 다양하고 시간이 오래 걸림ㅋ. 빠른 모델이 있는데 window를 지원 안함. linux 지원ㅋ
# wwwwwwwwwwww, lol
```
