### generater 살펴보기
- purpose: 효율적인 mpmery 사용
- 특징: iterator로써 반복기능 포함
- 문법: yield 키워드 함께 사용

```python
import time     # 차이점을 보기 위해서 일부러 시간간격으로 나누려고?
```

```python
# user define function
# function's function: return data
# function name: retrun_data

def return_data():
    data = []
    for ch in ['apple', 'kiwi', 'orange', 'banana']:
        data.append(ch)
        time.sleep(1)
        # 데이터를 1개 담는데 1초가 걸린다고 가정
        
    return data
```

```python
# user define function
# function's function: return data
# function name: retrun_data_yield

def return_data_yield():
    data = []
    for ch in ['apple', 'kiwi', 'orange', 'banana']:
        data.append(ch)
        time.sleep(1)
        # 데이터를 1개 담는데 1초가 걸린다고 가정
        yield data # 기능은 같은데 형식이 다름. generater
        
    return data
```

```python
return_data()
```

```python
return_data_yield()
```

- 
- 시간의 차이가 존재함

```python
for data in return_data():
    print(data)
```

```python
for data in return_data_yield():
    print(data)
```

- 
- 가져오는 것이 다름. 그치만 빨랐음.
- yield는 꺼내는 일단 주고 봄. 그래서 일반적인 iterater의 for문과 다르게 memory를 적게 차지함.
- 10,000 개의 데이터를 가져온다고 하면..

```python
# user define function
# function's function: return data
# function name: retrun_data_yield_2

def return_data_yield_2():
    data = ['apple', 'kiwi', 'orange', 'banana']
    
    time.sleep(1)
    # 데이터를 1개 담는데 1초가 걸린다고 가정
    yield from data # 기능은 같은데 형식이 다름. generater
```

```python
for data in return_data_yield_2():
    print(data)
```

- 1초 딜레이가 있지만 훨씬 빠름.
