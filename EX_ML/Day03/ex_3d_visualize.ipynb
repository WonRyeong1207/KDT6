#### [3D 시각화 예시]

```python
# 모듈로딩
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 데이터 생성
x = np.linspace(-10, 10, 11)
print(f"[x] shape: {x.shape}\n{x}\n")

X = np.tile(x, (11, 1))     # copy
print(f"[X] shape: {X.shape}\n{X}\n")

Y = np.transpose(X)
print(f"[Y] shape: {Y.shape}\n{Y}\n")

Z = np.random.rand(11, 11)
print(f"[Z] shape: {Z.shape}\n{Z}\n")
```

```python
# 시각화 진행
fig = plt.figure(figsize=(5,5))

# 3차원 창 설정 => projection='3d'
ax = fig.add_subplot(111, projection='3d')

# 평면 생성 : 2차원 어레이를 3차원 Surface로 표현
ax.plot_surface(X, Y, Z, color='yellow')
ax.set_zlim(-5, 5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')


plt.tight_layout()
plt.show()
```

- [예시2] 

```python
# 모듈로딩
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
## 임의의 데이터 
x = [0, 0, 1, 1]
y = [0, 1, 0, 1]
z = [-1, -1, -1, -1]

## 평면 좌표 데이터 생성
# - 좌표 범위 생성
X_data = np.arange(-0.5, 1.5, 0.1)
Y_data = np.arange(-0.5, 1.5, 0.1)

# - 행렬 ==> 좌표값 변환
X, Y = np.meshgrid(X_data, Y_data)
Z = (float(-5/7) * Y) + (float(-4/7) * X)
```

```python
## 시각화 
fig = plt.figure()
# projection : 객체가 그래프에 투영될 방법 설정
#              None, 'mollweide', 'polar', 'rectilinear', str
ax = fig.add_subplot(111, projection='3d')


# 데이터 출력
ax.plot(x, y, z, 
        linestyle="none", 
        marker="o", mfc="none", markeredgecolor="red")

# 데이터 평면 출력
# stride, cstride : row와 column에서 데이터에서 어느정도 graph 그리는데 사용할지 설정
# alpha : 그래프의 투명도  
# cmap : 색상 설정 color map


ax.plot_surface(X, Y, Z, 
                rstride=4, cstride=4, 
                alpha=0.4, cmap=cm.jet)

plt.show()
```
