- 4 - 3 Matplotlib으로 그래프 그리기

```python
# 데이터 준비
tips = sns.load_dataset("tips")
tips.head()
```

```python
# 일변량 그래프 그리기
#  히스토그램 그리기

plt.hist(tips['total_bill'], bins=10)     # bins : 구간을 나누는 속성

plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')

plt.show()
```

```python
# 책의 방법으로 그려보기

fig = plt.figure()

axes1 = fig.add_subplot(1, 1, 1)

axes1.hist(data=tips, x='total_bill', bins=10)

axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Frequency')

plt.show()
```

```python
# 이변량 그래프 그리기
# 산점도 그래프 그리기

plt.scatter(tips['total_bill'], tips['tip'])

plt.title('Scatterplot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

plt.show()
```

```python
# 책에 있는 코드

scatter_plot = plt.figure()
axes2 = scatter_plot.add_subplot(1, 1, 1)

axes2.scatter(tips['total_bill'], tips['tip'])

axes2.set_title('Scatterplot of Total Bill vs Tips')
axes2.set_xlabel('Total Bill')
axes2.set_ylabel('Tip')

plt.show()
```

```python
# 박스 그래프 그리기
# 여러개를 그리고 싶음. 데이터를 변수에 나눠서 저장
fe_tip = tips[tips['sex'] == 'Female']['tip']
ma_tip = tips[tips['sex'] == 'Male']['tip']

box_dict = plt.boxplot([fe_tip, ma_tip], tick_labels=['Female', 'Male'])

plt.title('Boxplot of Tips by Sex')
plt.xlabel('Sex')
plt.ylabel('Tip')
plt.show()
```

```python
# 책 코드

boxplot = plt.figure()

axes3 = boxplot.add_subplot(1, 1, 1)

axes3.boxplot(x=[tips[tips['sex'] == 'Female']['tip'],
                 tips[tips['sex'] == 'Male']['tip']], tick_labels=['Female', 'Male'])

axes3.set_title('Boxplot of Tips by Sex')
axes3.set_xlabel('Sex')
axes3.set_ylabel('Tip')

plt.show()
```
